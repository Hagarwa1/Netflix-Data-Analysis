**What is the maximum duration of a Movie/Show on Netflix?**

```python
data['Duration'].unique()
```

```python
# we can look for movies and TV shows separately

data[['Minutes', 'Unit']] = data['Duration'].str.split(' ', expand=True) # here duration column is split and stored into 2 new columns Minutes and Unit, splitting is done using ' ' operator
```

```python
# Now we will convert Minutes column which is in string to integer

data['Minutes_N']=pd.to_numeric(data['Minutes'],errors='coerce')
```

```python
#Maximum duration of the movie/show

data['Minutes_N'].max()
```

```python
# Name of the Movie/Show with max duration

data[data['Minutes_N']==data['Minutes_N'].max()]
```
